### Renombrar columnas del DataFrame

Hay dos maneras principales de renombrar columnas:

1.  **Usando el método `df.rename()` con un diccionario**: Esta es la forma recomendada cuando quieres cambiar solo algunas columnas.
2.  **Asignando una nueva lista a `df.columns`**: Útil cuando quieres renombrar todas las columnas a la vez, asegurándote de que la lista tenga el mismo número de elementos que las columnas actuales y en el orden correcto.

````python
# Para ver los nombres de las columnas actuales:
print("Nombres de columnas actuales:")
print(df.columns.tolist())

# Renombrar columnas específicas usando un diccionario
# Se reemplazará el nombre original por el nuevo nombre
new_column_names = {
    'Estimaciones de Tasa de Pobreza Multidimensional por Comuna, considerando Entorno y Redes y Aplicación de Metodologías de: Estimación directa, Estimación para Áreas Pequeñas (SAE) e Imputación de Medias por Conglomerados (IMC), 2015': 'Codigo Comuna',
    'Unnamed: 7': 'Metodologia Estimacion'
}

df = df.rename(columns=new_column_names)

# Mostrar los primeros 5 registros con las columnas renombradas para verificar
display(df.head())
````
